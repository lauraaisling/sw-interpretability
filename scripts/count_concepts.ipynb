## Count number of concepts found

```python
import os
import pandas as pd
import re
```

```python
dataset='ilsvrc12fine'
layer='Mixed_7b.cat_2'
SAVEFOLD0=f"../outputs/{dataset}"
SAVEFOLD=f"{SAVEFOLD0}/{layer}/"
```

Count concepts found starting from SVD directions: 

```python
evec_save = f'{SAVEFOLD}outputs/evecs/'
# evec_save # '../outputs/ilsvrc12fine/Mixed_7b.cat_2/outputs/evecs/'
```

```python
# append final_clusters for each neuron to one file
num_concepts = []
count_dirs_concepts = {}
for neuron_folder in next(os.walk(evec_save))[1]:
    neuron_dir = f'{evec_save}{neuron_folder}/'
#     print(neuron_dir)
    for filename in next(os.walk(neuron_dir))[2]:
        if filename.endswith(".txt") and re.match('final_clusters', filename):
#             print(filename) 
            file = f'{neuron_dir}{filename}'
            print(file)
            data = pd.read_csv(file, header=None, delimiter=r"\n")
            num_concepts.append(data.values)
            for val in data.values:
                if val[0]  in count_dirs_concepts:
                    count_dirs_concepts[val[0]] += 1
                else:
                    count_dirs_concepts[val[0]] = 1
num_concepts
```
